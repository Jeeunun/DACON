```python
sb.heatmap(df[['Gains', 'Losses', 'Dividends', 'Income']].corr(), annot=True, vmax=1, vmin=-1)
```

> Gains과 Losses : -0.013313 (약한 음의 상관관계) = Gains가 증가할 때 Losses가 감소하는 경향. 하지만 관련성이 낮다. <br>
> Gains과 Dividends : 0.081115 (약한 양의 상관관계) = Gains가 증가할 때 Dividends가 증가하는 경향. 하지만 관계성이 낮다. <br>
> Gains과 Income : 0.009555 (매우 약한 양의 상관관계) = Gains가 증가할 때 Income이 증가하는 경향. 하지만 관계성이 매우 낮다. <br>
> Losses과 Dividends : 0.015351 (매우 약한 양의 상관관계) = Losses가 증가할 때 Dividends가 증가하는 경향. 하지만 관계성이 매우 낮다. <br>
> Losses과 Income : 0.036502 (매우 약한 양의 상관관계) = Losses가 증가할 때 Income도 증가하는 경향. 하지만 관계성이 매우 낮다. <br>
> Dividends와 Income : 0.0024 (매우 약한 양의 상관관계) = Dividends가 증가할 때 Income도 증가하는 경향. 하지만 관계성이 매우 낮다. <br> 

> **0이 많은 Gains, Losses, Dividends는 Income에 유의미한 영향력을 미치지 못하는 것 같다. 0값이 많은 경우에 모델링에 좋은 영향을 미치지 못하기도 하고, 종속변수에 유의미한 영향을 미치지 못하니  삭제해도 좋을 것 같다 .**
